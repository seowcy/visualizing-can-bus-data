```python
import numpy as np
df["digital"] = np.where(df['diff']>=1.25, 0, 1)
```

```python
df["digital"].plot(figsize=(15,5))
```

```python
def myround(x, base=4):
    return base * round(x/base)

def dist_count_consecutive(l):
    result = []
    prev = l[0]
    count = 1
    time_delta = 0.0005
    for i in l[1:]:
        if i == prev:
            count += 1
        else:
            result.append((time_delta*myround(count), prev, myround(count)))
            count = 1
        prev = i
    return result
```

```python
results = dist_count_consecutive(df["digital"].values)
```

```python
sorted(results, key=lambda x: (x[0], x[1]))
```

```python
results
```

```python
one_zero = [j for sublist in [[i[1]] * int(i[2]/4) for i in results] for j in sublist]
```

```python
raw = ''.join([str(i) for i in one_zero])
raw
```

```python
messages = [i.strip('1') for i in raw.split('11111111') if i]
```

```python
messages
```

```python
def padhexa(s, padding=16):
    return s[2:].zfill(padding)

count = 1
for message in messages[:-1]:
    message_no_bs = message.replace("000001", "00000").replace("111110", "11111")
    msg_list = list(message_no_bs)
    start_bit = msg_list.pop(0)
    temp = []
    for i in range(11):
        temp.append(msg_list.pop(0))
    arb_id = ''.join(temp)
    rtr = msg_list.pop(0)
    ide = msg_list.pop(0)
    reserved = msg_list.pop(0)
    temp = []
    for i in range(4):
        temp.append(msg_list.pop(0))
    data_length = ''.join(temp)
    data_length_int = int(data_length, 2)
    temp = []
    for i in range(data_length_int * 8):
        temp.append(msg_list.pop(0))
    data = ''.join(temp)
    temp = []
    for i in range(16):
        temp.append(msg_list.pop(0))
    crc = ''.join(temp)
    ack = msg_list.pop(0)
    remainder = msg_list[:]
    
    print("------- message %s -------" % count)
    print("start_bit:", start_bit)
    print("arb_id:", arb_id, "(%s)" % hex(int(arb_id, 2))[2:])
    print("rtr:", rtr)
    print("ide:", ide)
    print("reserved:", reserved)
    print("data_length:", data_length, "(%s)" % hex(int(data_length, 2))[2:])
    data_hex = padhexa(hex(int(data, 2)), data_length_int*2)
    t = iter(data_hex)
    print("data:", data , "(%s)" % ' '.join(a+b for a,b in zip(t, t)))
    print("crc:", crc)
    print("ack:", ack)
    print("reminder:", remainder)
    print()
    count += 1
```
